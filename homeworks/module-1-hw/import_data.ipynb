```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# read in green tree taxi trip parquet data
green_taxi_trips = pd.read_parquet(
    'green_tripdata_2025-11.parquet',
    engine='fastparquet',
)
```

```python
green_taxi_trips.shape
```

```python
green_taxi_trips.head(5)
```

```python
green_taxi_trips.dtypes
```

```python
# create connection to db
engine = create_engine(f'postgresql://root:root@localhost:5432/ny_taxi')
```

```python
green_taxi_trips.head(0).to_sql(
    name='green_taxi_trips_11_2025',
    con=engine,
    if_exists='replace'
)
```

```python
green_taxi_trips.to_sql(
    name='green_taxi_trips_11_2025',
    con=engine,
    if_exists='append'
)
```
